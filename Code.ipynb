```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

## Used for plotting datasets with 3 variables
def plot_dataset(data,number):
    fig2 = px.scatter_3d(data, x = data.columns.values[1], y = data.columns.values[2], z = data.columns.values[3], color="Class", title = " 3D SCATTER PLOT OF DATA " + str(number) )
    fig2.show()

## Used for plotting datasets with 2 variables
def plot_2D_dataset(data,number):
    fig2 = px.scatter(data, x = data.columns.values[1], y = data.columns.values[2], color="Class", title = " 2D SCATTER PLOT OF DATA " + str(number))
    fig2.show()
  
## Used for clustering and plotting datasets with 3 variables
def clustering(data, algorithm):
    assert algorithm == 'kmeans' or algorithm == 'hierarchichal'    
    number_of_clusters = data.iloc[:, -1].nunique()
    data = data.iloc[:, 1:data.shape[1] - 1]
    if algorithm == 'kmeans':
        cluster = KMeans(n_clusters = number_of_clusters,random_state=1)
    else:
        cluster = AgglomerativeClustering(n_clusters = number_of_clusters,linkage = 'ward')
    d = cluster.fit(data)

    ## Plotting code goes here
    # Dendogram  
    if algorithm == 'hierarchichal':
        fig4 = plt.figure(figsize=(25, 18))
        dendo = sch.dendrogram(sch.linkage(data,method='ward'))
        fig4.show()  

    #Scatter Plot
    predicted_value = cluster.fit_predict(data)
    data['predicted_value'] = predicted_value
    fig = px.scatter_3d(data, x = data.columns.values[0], y = data.columns.values[1], z = data.columns.values[2], color="predicted_value", title = algorithm.upper() + " 3D SCATTER PLOT")
    if algorithm == 'kmeans':
      matrix = d.cluster_centers_
      fig.add_scatter3d(mode = 'markers',x = matrix[:,0],y = matrix[:,1],z = matrix[:,2],marker=dict(size=10, color="red"))
    fig.show()
    ##
    return cluster

## Used for clustering and plotting datasets with 2 variables
def clustering2D(data, algorithm):
    assert algorithm == 'kmeans' or algorithm == 'hierarchichal'
    number_of_clusters = data.iloc[:, -1].nunique()
    data = data.iloc[:, 1:data.shape[1] - 1]
    if algorithm == 'kmeans':
        cluster = KMeans(n_clusters = number_of_clusters,random_state=1)
    else:
        cluster = AgglomerativeClustering(n_clusters = number_of_clusters,linkage = 'ward')
    d = cluster.fit(data)

    ## Plotting code goes here
    if algorithm == 'hierarchichal':
        fig4 = plt.figure(figsize=(25, 18))
        dendo = sch.dendrogram(sch.linkage(data,method='ward'))
        fig4.show()

    #Scatter Plot
    predicted_value = cluster.fit_predict(data)
    data['predicted_value'] = predicted_value
    fig = px.scatter(data, x = data.columns.values[0], y = data.columns.values[1], color="predicted_value", title = algorithm.upper() + " 2D SCATTER PLOT")
    if algorithm == 'kmeans':
        matrix = d.cluster_centers_
        fig.add_scatter(mode = 'markers',x = matrix[:,0],y = matrix[:,1],marker=dict(size=10, color="red",))
    fig.show()
    ##
    return cluster
```

## Silhouette analysis to select optimum number of clusters for K means

#### For calculating the optimal number of clusters, we are doing silhouette analysis. Silhouette coefficient is calculated for each data point in the cluster. The formula to calculate silhouette coefficient for a single point is explained below:
#### First calculate a(i), that is that average distance of data point i from all other data points within same cluster

###$ a(i) = \frac{1}{|C_a|} \sum \limits _{j \in C_a,i \ne j} d(i,j)$

###$ b(i) = \min \limits _{C_a \ne C_b} \frac{1}{|C_b|} \sum \limits _{j \in C_b} d(i,j)$

###$s(i) = \frac{b(i)-a(i)}{\max \{a(i),b(i)\}}$

### The graphs shown below show the sihouette plots. In these plots the sihouette coefficient of every data point in the dataset is plottted. The x-axis has the silhouette score and the y axis shows cluster label. Following 2 criteria are used for finding optimal number of clusters:
###1) If for n_clusters = k, there is significant variation in the lengths of the silhouette coefficients for data points in different clusters, then such value of k is a bad choice.
###2) If for n_clusters = k, there are some clusters, whoes silhouette coefficients are below the average silhouette coefficient of the whole dataset, then such k is a bad choice.
####3) In case of ambiguity, we look width of the cluster which tells us the number of data points in each clusters. We choose n_clusters for which the width of all clusters are approximately same.

```python
K = [*range(2,8)]
for k in K:
  fig, (axis1) = plt.subplots(1)
  fig.set_size_inches(18,7)
  axis1.set_xlim([-0.1,1])
  axis1.set_ylim([0,data.shape[0] + (k+1)*10])


  cluster = KMeans(n_clusters=k,random_state=1)
  predicted_labels = cluster.fit(data).labels_
  average_silhouette_score = silhouette_score(data,predicted_labels)
  silhouette_score_per_data_point = silhouette_samples(data,predicted_labels)

  y_lower = 0

  for i in range(k):
    data_points_in_cluster_i = silhouette_score_per_data_point[predicted_labels == i]
    data_points_in_cluster_i.sort()
    y_upper = y_lower + data_points_in_cluster_i.shape[0]
    color = cm.nipy_spectral(float(i)/k)
    axis1.fill_betweenx(np.arange(y_lower,y_upper),0,data_points_in_cluster_i,facecolor=color,edgecolor=color,alpha=0.7)
    axis1.text(-0.05, y_lower + 0.5 * data_points_in_cluster_i.shape[0], str(i))
    y_lower = y_upper + 10

  axis1.set_title("The silhouette plot for " + str(k) +" clusters.")
  axis1.set_xlabel("The silhouette coefficient values")
  axis1.set_ylabel("Cluster label")
  axis1.axvline(x=average_silhouette_score, color="red", linestyle="--")
  axis1.set_yticks([])
  axis1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
  colors = cm.nipy_spectral(predicted_labels.astype(float) / k)
  print('The average silhouette score for ' + str(k) + ' cluster is = ' + str(average_silhouette_score))

plt.show()
```

## From the analysis of silhouette plots shown above n_clusters = 4,5,6,7 are bad choices due to variation in, lengths of the silhouette coefficients. It is dificult to decide between n_clusters = 2 and n_clusters = 3 because there is little variation in their lengths of silhouette coefficients

#  Calculating Internal Validation

## Internal validation is done using CH (Calinski Harabasz Score) and Dunn Index. Follwoing is the formula for CH Score and Dunn Index

#### Formula for CH Score
$ SSE(C_i) = \sum \limits _{x \in C_i} d(c_i,x)^2 = \frac{1}{2m_i} \sum \limits _{x \in C_i} \sum \limits _{y \in C_i} d(x,y)^{2} $
 
$SSB = \sum \limits ^{K} _{i=1}$ $m_id(c_i,c)^2 = \frac{1}{2K}\sum \limits ^{K} _{i=1} \sum \limits ^{K} _{j=1} \frac{m}{K} d(c_i,c_j)^2 $

$ C H = \frac{\frac{SSB_M}{M-1}}{\frac{SSE_M}{M}}$

#### Formula for Dunn Index


##$ DI_m = \frac{\min_{1 \le i \le j \le m} δ(C_i,C_j)}
{\max_{1 \le k \le m} Δ_k}$

### Following custom method calculates dunn index. The formula is min distance between any 2 cluster centroids, divided by, max distance between any 2 points in the cluster


```python
from scipy.spatial.distance import pdist

def dunn_index(cluster_centers,dataset):
  return np.min(pdist(cluster_centers))/np.max(pdist(dataset))
```

### For calculating CH score we have used method from sklearn.metrics

```python
from sklearn.metrics import calinski_harabasz_score
def get_cluster_centers_for_hierarchical(original_dataframe,scaled_data_array,number_of_clusters):
  cluster_centers = []
  original_dataframe_copy = original_dataframe.copy()
  original_dataframe_copy['cluster'] = AgglomerativeClustering(n_clusters=number_of_clusters,linkage='ward').fit_predict(scaled_data_array)
  for cluster_label in original_dataframe_copy.cluster.unique():
    cluster_data_points = scaled_data_array[original_dataframe_copy.cluster == cluster_label]
    cluster_centers.append(np.mean(cluster_data_points,axis=0))
  return cluster_centers
```

```python
kmeans_clustering = KMeans(n_clusters=2,random_state=1).fit(data)
print('CH Score for K-means with k=2: ' + str(calinski_harabasz_score(data,kmeans_clustering.labels_)))
print('Dunn Index for K-means with k=2: ' + str(dunn_index(kmeans_clustering.cluster_centers_,data)) + '\n\n')


kmeans_clustering = KMeans(n_clusters=3,random_state=1).fit(data)
print('CH Score for K-means with k=3: ' + str(calinski_harabasz_score(data,kmeans_clustering.labels_)))
print('Dunn Index for K-means with k=3: ' + str(dunn_index(kmeans_clustering.cluster_centers_,data)) + '\n\n')

heirarchical_clustering = AgglomerativeClustering(n_clusters=2,linkage='ward').fit(data)
print('CH Score for Heirarchical with number of clusters=2: ' + str(calinski_harabasz_score(data,heirarchical_clustering.labels_)))
cluster_centroids = get_cluster_centers_for_hierarchical(world_df,data,2)
print('Dunn Index for heirarchical with number of clusters=2: ' + str(dunn_index(cluster_centroids,data)) + '\n\n')

heirarchical_clustering = AgglomerativeClustering(n_clusters=3,linkage='ward').fit(data)
print('CH Score for Heirarchical with number of clusters=3: ' + str(calinski_harabasz_score(data,heirarchical_clustering.labels_)))
cluster_centroids = get_cluster_centers_for_hierarchical(world_df,data,3)
print('Dunn Index for heirarchical with number of clusters=3: ' + str(dunn_index(cluster_centroids,data)))
```

### In both kmeans and hierarchical, n_cluster = 2 provides higher internal validation metric as compared to n_cluster = 3. Based on these evidences, n_cluster = 2 would be better as compared to n_cluster = 3
